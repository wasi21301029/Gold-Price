```python
import matplotlib.pyplot as plt
import pandas as pd
from pandas import DataFrame
import seaborn as sns
```

```python
data = pd.read_csv('GoldUP.csv')
```

```python
data.head()
```

```python
t = []
g = []
u = []
for i in range(len(data["Date"])):
    a = int(data["Date"][i][6:10])
    b = int(data["Gold_Price"][i])
    c = int(data["USD_INR"][i])
    if a not in t:
        t.append(a)
        g.append(b)
        u.append(c)
```

```python
#Increase of gold price in yearly basis.

plt.figure(figsize=(10,8)) 
plt.plot(t, g)
plt.xlabel("Date")
plt.ylabel("Price")
plt.show()
```

```python
plt.figure(figsize=(10,8)) 
plt.plot(t, u)
plt.xlabel("Date")
plt.ylabel("Price")
plt.show()
```
